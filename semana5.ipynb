```
def largoTexto(texto):
    largo = 0
    for elemento in texto:
        largo += 1
    return largo
'''
nombre: coincidenciasEnArchivos
entradas: dos; nombreArchivo, textoABuscar
salidas: numero de coincidencias de textoABuscar en textoBase
restricciones: textoBase y textoABuscar deben ser str
    textoBase y textoABuscar no pueden estar vacios
'''
def coincidenciasEnArchivos(nombreArchivo, textoABuscar):
    if isinstance(nombreArchivo, str) and isinstance(textoABuscar, str):
        archivo = open(nombreArchivo, "r")
        largo = largoTexto(textoABuscar)
        leer = archivo.readlines()
        coincidencias = 0
        for linea in leer:
            while linea != "":
                if linea[0: largo] == textoABuscar:
                    coincidencias += 1
                    linea = linea[1:]
                else:
                    linea = linea[1:]
        return coincidencias
    else:
        return "Error: los parametros deben ser str"

'''
nombre: eliminarCoincidencias
entradas: dos; textoBase, textoABuscar
salidas: textoBase con todas las coincidencias en textoAbuscar eliminadas
restricciones: textoABuscar y textoBase deben ser str
        textoABuscar y textoBase no deben estar vacions
'''
def eliminarCoincidencias(textoBase, textoABuscar):
    if(isinstance(textoBase, str) and isinstance(textoABuscar, str)):
        if(textoBase != "" and textoABuscar != ""):
            largo = largoTexto(textoABuscar)
            final = ""
            while(textoBase != ""):
                if(textoBase[:largo] == textoABuscar):
                    textoBase = textoBase[largo:]
                else:
                    final += textoBase[0]
                    textoBase = textoBase[1:]
            return final
        else:
            return "Error: los valores no pueden estar vacios"
    else:
        return "Error: las entradas deben ser str"

'''
nombre: reemplazarCoincidencias
entradas: tres; extoBase, textoABuscar, reemplazo
salida: reeplaza textoABuscar en textoBase con reemplazo
restricciones: reemplazo, textoABuscar y textoBase deben ser str
        reemplazo, textoABuscar y textoBase no deben estar vacios
'''
def reemplazarCoincidencias(textoBase, textoABuscar, reemplazo):
    if(isinstance(textoBase, str) and isinstance(textoABuscar, str) and isinstance(reemplazo, str)):
        if(textoBase != "" and textoABuscar != "" and reemplazo != ""):
            largo = largoTexto(textoABuscar)
            final = ""
            while(textoBase != ""):
                if(textoBase[:largo] == textoABuscar):
                    textoBase = textoBase[largo:]
                    final += reemplazo
                else:
                    final += textoBase[0]
                    textoBase = textoBase[1:]
            return final
        else:
            return "Error: los valores no pueden estar vacios"
    else:
        return "Error: las entradas deben ser str"

'''
nombre: totalDigitosDivisible
entradas: dos; num, divisor
salida: numero de digitos en num divisibles
restricciones: num y divisores deben ser numeros enteros mayores a 0
'''
def totalDigitosDivisible(num, divisor):
    #num y divisor deben ser numeros enteros
    if(isinstance(num, int) and isinstance(divisor, int)):
        if((num > 0) and (divisor > 0)):
            #resultado cuneta todos los numeros divisibles
            resultado = 0
            while(num != 0):
                #revisa si el primer digito de num es divisible
                revisar = num % 10
                if((revisar % divisor) == 0):
                    resultado += 1
                    num = num // 10
                else:
                    num = num // 10
            return resultado
        else:
            return "Error: los valores deben der mayores a 0"
    else:
        return "Error: los valores deben ser numeros enteros"

'''
nombre: totalNumerosPrimos
entradas: una; num
salidas: cantidad de numeros primos desde 1 a num
restricciones: num debe ser un numero positivo mayor a 0
'''
def totalNumerosPrimos(num):
    if(isinstance(num, int)):
        if(num > 0):
            resultado = 0
            while(num != 1):
                revisar = 1
                contador = 0
                while(revisar != num):
                    if(num % revisar == 0):
                        contador += 1
                        revisar += 1
                    else:
                        revisar += 1
                if(contador >= 2):
                    num -= 1
                else:
                    num -= 1
                    resultado += 1
            return resultado
        else:
            return "Error: num debe ser un numero mayor a 0"
    else:
        return "Error: num debe ser un numero entero"

'''
nombre: sumatoria
entradas: dos; i, n
salidas: sumatoria de i a n
restricciones: i y n deben ser numero enteros positivos
    i debe ser menor a n
    n debe ser mayor a 0
'''
def sumatoria(i, n):
    if(isinstance(i, int) and isinstance(n, int)):
        if(n > i and n > 0):
            resultado = 0
            for numero in range(i, n + 1):
                resultado += numero
            return resultado
        else:
            return "Error: n debe ser mayor a i y 0"
    else:
        return "Error: i y n deben ser numeros entero positivos"

def multi_for(lista):
    if(isinstance(lista, list)):
        if(lista != []):
            resultado = 1
            for elemento in lista:
                resultado *= elemento
            return resultado
        else:
            return "Error: La lista no debe ser nula"
    else:
        return "Error: debe ingresar una lista"

def multi_while(lista):
    if(isinstance(lista, list)):
        if(lista != []):
            resultado = 1
            while(lista != []):
                multi = lista[0]
                resultado *= multi
                lista = lista[1:]
            return resultado
        else:
            return "Error: La lista no debe ser nula"
    else:
        return "Error: debe ingresar una lista"

def largoTexto(texto):
    contar = -1
    for elemento in texto:
        contar += 1
    return contar

def buscarEnTexto(textoBase, textoABuscar):
    if(isinstance(textoBase, str) and isinstance(textoABuscar, str)):
        if(textoBase != "" and textoABuscar != ""):
            resultado = False
            contar = largoTexto(textoABuscar)
            while(textoBase != ""):
                if(textoABuscar == textoBase[0: contar + 1]):
                    resultado = True
                    textoBase = textoBase[1:]
                else:
                    textoBase = textoBase[1:]
            return resultado
        else:
            return "Error: las entradas no deben estar vacias"
    else:
        return "Error: las entradas deben ser de tipo str"

def buscarEnTexto_v2(textoBase, textoABuscar):
    if(isinstance(textoBase, str) and isinstance(textoABuscar, str)):
        if(textoBase != "" and textoABuscar != ""):
            resultado = False
            contar = largoTexto(textoABuscar)
            while(textoBase != ""):
                if(textoABuscar.upper() == textoBase[0: contar + 1].upper()):
                    resultado = True
                    textoBase = textoBase[1:]
                else:
                    textoBase = textoBase[1:]
            return resultado
        else:
            return "Error: las entradas no deben estar vacias"
    else:
        return "Error: las entradas deben ser de tipo str"

def coincidenciasEnTexto(textoBase, textoABuscar):
    if(isinstance(textoBase, str) and isinstance(textoABuscar, str)):
        if(textoBase != "" and textoABuscar != ""):
            resultado = 0
            contar = largoTexto(textoABuscar)
            while(textoBase != ""):
                if(textoABuscar.upper() == textoBase[0: contar + 1].upper()):
                    resultado += 1
                    textoBase = textoBase[1:]
                else:
                    textoBase = textoBase[1:]
            return resultado
        else:
            return "Error: las entradas no deben estar vacias"
    else:
        return "Error: las entradas deben ser de tipo str"

def coincidenciasEnArchivo(nombreArchivo, textoABuscar):
    archivo = open(nombreArchivo, "r")
    contenido = archivo.readlines()
    resultado = 0
    contar = largoTexto(textoABuscar)
    for linea in contenido:
        while (linea != ""):
            if (textoABuscar.upper() == linea[0: contar + 1].upper()):
                resultado += 1
                linea = linea[1:]
            else:
                linea = linea[1:]
    return resultado
```
