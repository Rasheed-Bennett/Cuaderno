###Codigo

```
'''
nombre: esPar
entrada: una; num
salidas: si num es un numero par: True
        si num es un numero impar: False
restricciones:num debe ser mayor o igual a cero
            num debe ser un numero entero
'''
def esPar(num):
    #num debe ser un numero entero
    if(isinstance(num, int)):
        #num debe ser un numero mayor o igual a 0
        if(num >= 0):
            #si num es par devolver True
            if(num % 2 == 0):
                return True
            else:
                return False
        else:
            return "Error: num debe ser un numero mayor o igual a 0"
    else:
        return "Error: debe ingresar un numero entero"
```

```
'''
nombre:contarDigitosPares
entradas: una; num
salidas: cantidad de digitos pares en num
restricciones: se debe usar while o for
				num debe ser un numero entero
'''
def contarDigitosPares(num):
    #Num debe ser un numero entero y mayor o igual a cero
    if(isinstance(num, int)):
        if(num >= 0):
            #resultado contara el numero de digitos
            resultado = 0
            while(num >= 10):
                #pares confirmara si el primer digito es par
                pares = num % 10
                #Elimina el primer digito de num
                #Si num es un solo digito no eliminara num
                num = num // 10
                if(pares % 2 == 0):
                    resultado += 1
                else:
                    continue
            #revisa el ultimo digito se num
            if(num % 2 == 0):
                resultado += 1
                return resultado
            else:
                return resultado
        else:
            return "Error: num debe ser un numero positvo"
    else:
        return "Error: debe ingresar un numero entero"
```

##Ejercicio 13

###factorial
Diagrama

image.png

###Pseudocodigo

```
Algoritmo factorial
	resultado = 1
	Leer num
	Si 0 <= num Entonces
		Para numero<-1 Hasta num Hacer
			resultado = numero * resultado
		FinPara
		Escribir resultado
	SiNo
		Escribir "Error: num debe ser mayor o igual a cero"
	FinSi
FinAlgoritmo
```
###Codigo
```
'''
nombre: factorial
entradas: una; num
salidas: factorial de num
restricciones: num debe ser un numero entero positivo
    usar while o for
'''
def factorial(num):
    #num debe ser un numero entero positivo
    if(isinstance(num, int)):
        if(num >= 0):
            #mostrara el resultado del factorial
            resultado = 1
            #multiplica todos los numero entre 1 y num
            for numero in range(1, num + 1):
                resultado *= numero
            return resultado
        else:
            return "Error: debe ingresar un numero entero positivo"
    else:
        return "Error: debe ingresar un numero entero"
```

##Ejercicio 14
###grupo2Digitos
Diagrama

image.png

###Pseudocodigo

```
Algoritmo grupo2Digitos
	resultado = 0
	Leer num
	Si num >= 0 Entonces
		Mientras num >= 10 Hacer
			resultado = resultado + 1
			num = num / 100
		FinMientras
		Escribir resultado
	SiNo
		Escribir "Error: num debe ser un numero entero positvo"
	FinSi
FinAlgoritmo
```

###Codigo

```
'''
nombre: grupo2Digitos
entradas: una; num
salida: numero de grupos de dos digitos en num
restricciones: num debe ser un numero entero positivo
    debe usar while o for
'''
def grupo2Digitos(num):
    if(isinstance(num, int) and num >= 0):
        #contara todos los grupos de 2 digitos
        resultado = 0
        while(num >= 10):
            resultado += 1
            #Elimina los primeros dos digitos que ya fueron contados en cada ciclo
            num = num // 100
        return resultado
    else:
        return "Error: debe ingresar un numero entero positivo"
```
