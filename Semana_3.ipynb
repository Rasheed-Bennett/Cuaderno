codigo
```
'''
nombre: esPrimo
entrada = una; num
salida: verdadero si num es primo
        falso si num no es primo
restricciones: num debe ser un numero entero positivo
'''
def esPrimo(num):
    if(isinstance(num, int) and num > 0):
        if(num > 1):
            primo = 1
            while(primo != num):
                if(num % primo == 0 and primo != 1):
                    return False
                else:
                    primo += 1
            return True
        else:
            return False
    else:
        return "Error: la variable num debe ser un numero entero positivo mayor a cero"
```
