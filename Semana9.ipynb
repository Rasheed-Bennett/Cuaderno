### ValidaVector

####Codigo
```
'''
nombre: validaVector
entradas: una; lista
salidas: True si la lista es un vertor cuyos numeros sean enteros
    False si la lista no es un vector con numeros enteros
restricciones: lista debe ser una lista
    lista debe contener numeros enteros
'''
def validaVector(lista):
    if isinstance(lista, list):
        for elementos in lista:
            if isinstance(lista, int):
                continue
            else:
                return False
        return True
    else:
        return "Error: debe ingresar una lista"
```

### SumaVector

#### Codigo

```
'''
nombre: sumaVector
entradas: dos; lista1, lista2
salidas: la suma de las dos listas
restricciones: lista1 y lista2 deben contener numeros enteros
    los vectores deben tener el mismo tamaño
'''
def sumaVectores(lista1, lista2):
    if isinstance(lista1, list) and isinstance(lista2, list):
        for elementos in lista1:
            if isinstance(elementos, int):
                continue
            else:
                return "Error: los elementos de lista1 deben ser numeros enteros"
        for elementos in lista2:
            if isinstance(elementos, int):
                continue
            else:
                return "Error: los elementos de lista2 deben ser numeros enteros"
        return sumaVectores_Aux(lista1, lista2)
    else:
        return "Error: las entradas debens ser listas"

def sumaVectores_Aux(lista1, lista2):
    resultado = []
    finalizar = False
    contar = 0
    while finalizar == False:
        try:
            lista1sum = lista1[contar]
            try:
                lista2sum = lista2[contar]
                resultado += [lista2sum + lista1sum]
                contar += 1
            except:
                return "Error: los vectores deben tener el mismo tamaño"
        except:
            try:
                lista2sum = lista2[contar]
                return "Error: los vectores deben tener el mismo tamaño"
            except:
                finalizar = True
    return resultado
```

### ProductoEscalar

#### Codigo
```
def productoEscalar(vector1, vector2):
    if isinstance(vector1, list) and isinstance(vector2, list):
        for elementos in vector1:
            if isinstance(elementos, int):
                continue
            else:
                return "Error: los elementos de lista1 deben ser numeros enteros"
        for elementos in vector2:
            if isinstance(elementos, int):
                continue
            else:
                return "Error: los elementos de lista2 deben ser numeros enteros"
        return productoEscalar_Aux(vector1, vector2)
    else:
        return "Error: las entradas debens ser listas"

def productoEscalar_Aux(vector1, vector2):
    resultado = 0
    finalizar = False
    contar = 0
    while finalizar == False:
        try:
            lista1sum = vector1[contar]
            try:
                lista2sum = vector2[contar]
                resultado += lista2sum * lista1sum
                contar += 1
            except:
                return "Error: los vectores deben tener el mismo tamaño"
        except:
            try:
                lista2sum = vector2[contar]
                return "Error: los vectores deben tener el mismo tamaño"
            except:
                finalizar = True
    return resultado
```

### vectorXEscalar

#### Codigo
```
'''
nombre: vectorXEscalar
entradas: dos; vector, num
salidas: lista del vector encalado por num
restricciones: vector debe ser una lista con numeros enteros
    num debe ser un numero entero
'''
def vectorXEscalar(vector, num):
    if isinstance(vector, list):
        if isinstance(num, int):
            for elementos in vector:
                if isinstance(elementos, int):
                    continue
                else:
                    return "Error: los elementos de lista1 deben ser numeros enteros"
            return vectorXEscalar_Aux(vector, num)
        else:
            return "Error: num debe ser un numero entero"
    else:
        return "Error: las entradas debens ser listas"

def vectorXEscalar_Aux(vector, num):
    resultado = []
    finalizar = False
    contar = 0
    while finalizar == False:
        try:
            lista1sum = vector[contar]
            resultado += [lista1sum * num]
            contar += 1
        except:
            finalizar = True
    return resultado
```
